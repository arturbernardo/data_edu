```python
df_rs_poa = df[(df['SG_UF'] == 'RS') & (df['NO_MUNICIPIO'] == 'Lajeado')]
```

```python
df_rs_poa
```

```python
import requests

API_KEY = ''

def get_coordinates(address, api_key):
    url = "https://maps.googleapis.com/maps/api/geocode/json"
    params = {
        'address': address,
        'key': api_key
    }
    response = requests.get(url, params=params)
    data = response.json()

    if data['status'] == 'OK':
        location = data['results'][0]['geometry']['location']
        return location['lat'], location['lng']
    else:
        print(f"Erro: {data['status']}")
        return None, None
```

```python
# df_rs_poa[''] NO_ENTIDADE DS_ENDERECO NU_ENDERECO
```

```python
latitudes = []
longitudes = []

for idx, row in df_rs_poa.iterrows():
    endereco = "Lajeado, Rio Grande do Sul, rua " + row['DS_ENDERECO'] + ", numero " + row['NU_ENDERECO']
    lat, lng = get_coordinates(endereco, API_KEY)

    latitudes.append(lat)
    longitudes.append(lng)

    time.sleep(0.2)  # Pausa para evitar limites de rate (200ms entre chamadas)

# Adiciona ao DataFrame
df_rs_poa['latitude'] = latitudes
df_rs_poa['longitude'] = longitudes
```

```python
import folium

if df_rs_poa['latitude'].notnull().any():
    lat0 = df_rs_poa['latitude'].dropna().iloc[0]
    lon0 = df_rs_poa['longitude'].dropna().iloc[0]
else:
    lat0, lon0 = -29.4685, -51.9653

mapa = folium.Map(location=[lat0, lon0], zoom_start=5)

for idx, row in df_rs_poa.iterrows():
    lat = row['latitude']
    lon = row['longitude']
    endereco = row['NO_ENTIDADE']

    if pd.notnull(lat) and pd.notnull(lon):
        folium.Marker(
            location=[lat, lon],
            popup=endereco
        ).add_to(mapa)
```
